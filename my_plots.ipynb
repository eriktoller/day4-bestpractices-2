```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')

n = 100
a = np.random.random(n)
x = np.arange(n)/n

plt.plot(x, a, 'o', color='black');
```

```python
num = 10000
mu, sigma = 0, 0.1 # mean and standard deviation
xx = np.random.normal(mu, sigma, num)
mu, sigma = .1, 0.2 # mean and standard deviation
yy = np.random.normal(mu, sigma, num)

fig, ((ax0, ax1),(ax2, ax3)) = plt.subplots(nrows=2, ncols=2)
ax0.hist(xx, bins = 100, color='blue');
ax1.hist(yy, bins = 100, color='green');
ax2.hist(xx, bins = 100, color='blue',alpha = 0.5);
ax2.hist(yy, bins = 100, color='green',alpha = 0.5);
ax3.hist(xx, bins = 100, color='blue',histtype='step');
ax3.hist(yy, bins = 100, color='green',histtype='step');
```
